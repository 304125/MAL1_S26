## Linear Regression - manually
Examples come from the course book Hands-On Machine Learning with Scikit-Learn, Keras & TensorFlow, A. Geron, 3rd edition.
Repository at: https://github.com/ageron/handson-ml3/tree/main#

```python
import numpy as np

np.random.seed(42)  # to make this code example reproducible
m = 100  # number of instances
X = 2 * np.random.rand(m, 1)  # column vector
y = 4 + 3 * X + np.random.randn(m, 1)  # column vector, target variable
X
```

```python
y
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.plot(X, y, "b.")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 2, 0, 15])
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import add_dummy_feature

# add a column of ones to X to account for the intercept term (theta_0)
X_b = add_dummy_feature(X)
theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
```

```python
theta_best # finds the best values for theta_0 and theta_1 given our data
```

```python
X_new = np.array([[0], [2]]) # 2 new observations that I want to predict for
X_new_b = add_dummy_feature(X_new)  # add x0 = 1 to each instance
X_new_b
```

```python
# make predictions on new data using the learned coefficients
y_predict = X_new_b @ theta_best
y_predict
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))  # extra code – not needed, just formatting
plt.plot(X_new, y_predict, "r-", label="Predictions")
plt.plot(X, y, "b.")

# extra code – beautifies and saves Figure 4–2
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 2, 0, 15])
plt.grid()
plt.legend(loc="upper left")
plt.show()
```

## Linear regression - using Scikit-Learn

```python
from sklearn.linear_model import LinearRegression

# training the model - use the dataset without column of ones, since Scikit-Learn's LinearRegression will add the intercept automatically
lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
# making predictions using the trained model
lin_reg.predict(X_new)
```

## Polynomial regression

```python
np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X ** 2 + X + 2 + np.random.randn(m, 1)
X
```

```python
y
```

```python
# extra code – this cell generates and saves Figure 4–12
plt.figure(figsize=(6, 4))
plt.plot(X, y, "b.")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([-3, 3, 0, 10])
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X) # this creates new features x1^2, x1^3, ... up to the specified degree
```

```python
X[0]
```

```python
X[0] ** 2, X_poly[0]
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
X_new = np.linspace(-3, 3, 100).reshape(100, 1)
X_new_poly = poly_features.transform(X_new)
y_new = lin_reg.predict(X_new_poly)

plt.figure(figsize=(6, 4))
plt.plot(X, y, "b.")
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.legend(loc="upper left")
plt.axis([-3, 3, 0, 10])
plt.grid()
# plt.show()
```

```python
# underfitting and overfitting

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

plt.figure(figsize=(6, 4))

for style, width, degree in (("r-", 2, 1), ("y-", 2, 2), ("g-", 1, 300)):
    polybig_features = PolynomialFeatures(degree=degree, include_bias=False)
    std_scaler = StandardScaler()
    lin_reg = LinearRegression()
    polynomial_regression = make_pipeline(polybig_features, std_scaler, lin_reg)
    polynomial_regression.fit(X, y)
    y_newbig = polynomial_regression.predict(X_new)
    label = f"{degree} degree{'s' if degree > 1 else ''}"
    plt.plot(X_new, y_newbig, style, label=label, linewidth=width)

plt.plot(X, y, "b.", linewidth=3)
plt.legend(loc="upper left")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([-3, 3, 0, 10])
plt.grid()
plt.show()
```

## Regularized linear models

```python
# generating a noisy dataset
np.random.seed(42)
m = 20
X = 3 * np.random.rand(m, 1)
y = 1 + 0.5 * X + np.random.randn(m, 1) / 1.5
X_new = np.linspace(0, 3, 100).reshape(100, 1)
```

```python
# a quick peek at the dataset we just generated
plt.figure(figsize=(6, 4))
plt.plot(X, y, ".")
plt.xlabel("$x_1$")
plt.ylabel("$y$  ", rotation=0)
plt.axis([0, 3, 0, 3.5])
plt.grid()
plt.show()
```

```python
#linear regression model without regularization, ridge, lasso and elastic net regression models
from sklearn.linear_model import Ridge, Lasso, ElasticNet

lin_reg = LinearRegression()
ridge_reg = Ridge(alpha=1)
lasso_reg = Lasso(alpha=0.1)
elastic_net_reg = ElasticNet(alpha=0.1, l1_ratio=0.5)

lin_reg.fit(X, y)
ridge_reg.fit(X, y)
lasso_reg.fit(X, y)
elastic_net_reg.fit(X, y)

y_lin_reg = lin_reg.predict(X_new)
y_ridge = ridge_reg.predict(X_new)
y_lasso = lasso_reg.predict(X_new)
y_elastic_net = elastic_net_reg.predict(X_new)

plt.figure(figsize=(6, 4))
plt.plot(X, y, "b.", linewidth=3)
plt.plot(X_new, y_lin_reg, "b-", label="Linear Regression", linewidth=2)
plt.plot(X_new, y_ridge, "r-", label="Ridge", linewidth=2)
plt.plot(X_new, y_lasso, "g-", label="Lasso", linewidth=2)
plt.plot(X_new, y_elastic_net, "y-", label="Elastic Net", linewidth=2)
plt.legend(loc="upper left")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 3, 0, 3.5])
plt.grid()
plt.show()
```

```python
# impact of alpha hyperparameter on Ridge regression

plt.figure(figsize=(6, 4))
for alpha, style in ((0.1, "r-"), (1, "g-"), (10, "y-")):
    ridge_reg = Ridge(alpha=alpha)
    ridge_reg.fit(X, y)
    y_ridge = ridge_reg.predict(X_new)
    label_ridge = f"Ridge alpha={alpha}"
    plt.plot(X_new, y_ridge, style, label=label_ridge, linewidth=2)

plt.plot(X, y, "b.", linewidth=3)
plt.legend(loc="upper left")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 3, 0, 3.5])
plt.grid()
plt.show()
```

```python
# impact of alpha hyperparameter on Lasso regression

plt.figure(figsize=(6, 4))
for alpha, style in ((0.1, "r-"), (1, "g-"), (10, "y-")):
    lasso_reg = Lasso(alpha=alpha)
    lasso_reg.fit(X, y)
    y_lasso = lasso_reg.predict(X_new)
    label_lasso = f"Lasso alpha={alpha}"
    plt.plot(X_new, y_lasso, style, label=label_lasso, linewidth=2)

plt.plot(X, y, "b.", linewidth=3)
plt.legend(loc="upper left")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 3, 0, 3.5])
plt.grid()
plt.show()
```

```python
# impact of alpha and l1_ratio hyperparameters on Elastic Net regression
# l1=1 corresponds to Lasso regression, l1=0 corresponds to Ridge regression, and values in between correspond to a combination of both
plt.figure(figsize=(6, 4))
for l1_ratio, style in ((0, "r-"), (0.3, "g-"), (0.5, "y-"), (0.8, "m-"), (1, "c-")):
    elastic_net_reg = ElasticNet(alpha=1, l1_ratio=l1_ratio)
    elastic_net_reg.fit(X, y)
    y_elastic_net = elastic_net_reg.predict(X_new)
    label_elastic_net = f"Elastic Net alpha={1} l1_ratio={l1_ratio}"
    plt.plot(X_new, y_elastic_net, style, label=label_elastic_net, linewidth=2)

plt.plot(X, y, "b.", linewidth=3)
plt.legend(loc="upper left")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 3, 0, 3.5])
plt.grid()
plt.show()
```
